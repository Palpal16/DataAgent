```python
import pandas as pd

# Read the sweep results CSV
# add level of difficulty
data = pd.read_csv('../output_google_colab/sweep_results.csv')
data.head(5)
```

```python
data.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot by level of difficulty, grouped by best_n
sns.barplot(x='difficulty', y='csv_score', hue='best_of_n', data=data)
```

```python
import seaborn as sns

sns.catplot(
    data=data,
    kind="bar",
    x="difficulty",
    y="csv_score",
    hue="best_of_n",
    col="two_stage_cot",
)
```
